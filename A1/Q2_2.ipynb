```python
import pandas as pd
import numpy as np

df = pd.read_csv('Heart.csv')
df = df.dropna()
df = df.drop(['Unnamed: 0'], axis=1)
df.shape
```

```python
df.shape
```

```python
df['ChestPain'] = df['ChestPain'].map({'typical':0, 'asymptomatic':1, 'nonanginal':2, 'nontypical':3})
df['Thal'] = df['Thal'].map({'fixed':0, 'normal':1, 'reversable':2})
df['AHD'] = df['AHD'].map({'Yes': 1, 'No': 0})
```

```python
df
```

```python
df_c1 = df[df.AHD == 1]
df_c2 = df[df.AHD == 0]

df_c1_x = df_c1.drop(columns=['AHD'])
df_c2_x = df_c2.drop(columns=['AHD'])

df_c1_y = df_c1['AHD']
df_c2_y = df_c2['AHD']
```

```python
m1 = df_c1_x.mean()
m1 = m1.to_numpy()
m1
```

```python
m2 = df_c2_x.mean()
m2 = m2.to_numpy()
m2
```

```python
x1 = df_c1_x.to_numpy()
x1
```

```python
x2 = df_c2_x.to_numpy()
x2
```

```python
y1 = df_c1_y.to_numpy()
y1
```

```python
y2 = df_c2_y.to_numpy()
y2
```

FDA

```python
s1 = (1/(len(x1)))*np.matmul((x1-m1).T, (x1-m1))
s1.shape
```

```python
s2 = (1/(len(x2)))*np.matmul((x2-m2).T, (x2-m2))
s2.shape
```

```python
n1 = len(x1)
n2 = len(x2)
print(n1, n2)
# omega = ()
```

```python
omega = np.matmul(np.linalg.inv(n1*s1 + n2*s2), (m1-m2))
```

```python
omega
```

```python
px1 = np.matmul(x1, omega)
print(px1)
```

```python
px2 = np.matmul(x2, omega)
print(px2)
```

```python
def sig(x):
    return 1/(1+np.exp(-x))
```

```python
px = []
py = []
for i in px1:
    px.append([1,i])
for i in y1:
    py.append(i)
for i in px2:
    px.append([1,i])
for i in y2:
    py.append(i)
px = np.array(px)
py = np.array(py)
print(px)
print(py)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(px, py, test_size=0.1, random_state=42)
```

```python
weight = np.zeros(2)
print(weight.reshape(-1,1))
```

```python
max_iter = 500000
learning_rate = 0.1
weight = np.zeros(2)

for _ in range(max_iter):
    p = sig(np.dot(x_train, weight))
    weight = weight - (learning_rate/(n1+n2))*np.matmul(x_train.T, (p-y_train))

print(weight)
```

```python
y_pred = sig(np.dot(x_test, weight))
y_final = [1 if i > 0.5 else 0 for i in y_pred]
y_final
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_final))
```

PCA+FDA

```python
from sklearn.decomposition import PCA
import numpy as np

pca = PCA(n_components=13)

x = np.concatenate((x1, x2), axis=0)
y = np.concatenate((y1, y2), axis=0)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=42)
pca.fit(x_train)
x_train = pca.transform(x_train)

print(x_train)
```

```python
c0_x = [x_train[i] for i in range(len(x_train)) if y_train[i] == 1]
c0_x = np.array(c0_x)
c0_y = [i for i in y_train if i == 1]
print(c0_x)
print(c0_y)
```

```python
c1_x = [x_train[i] for i in range(len(x_train)) if y_train[i] == 0]
c1_x = np.array(c1_x)
c1_y = [i for i in y_train if i == 0]
print(c1_x)
print(c1_y)
```

```python
m0 = np.mean(c0_x, axis=0)
m0
```

```python
m1 = np.mean(c1_x, axis=0)
m1
```

```python
s0 = (1/(len(c0_x)))*np.matmul((c0_x-m0).T, (c0_x-m0))
s0
```

```python
s1 = (1/(len(c1_x)))*np.matmul((c1_x-m1).T, (c1_x-m1))
s1
```

```python
n0 = len(c0_x)
n1 = len(c1_x)

omega = np.matmul(np.linalg.inv(n0*s0 + n1*s1), (m0-m1))
```

```python
omega
```

```python
px = np.dot(omega, x_train.T)
px
```

```python
px_train = [[1, i] for i in px]
px_train = np.array(px_train)
px_train
```

```python
weight = np.zeros(2)
print(weight.reshape(-1, 1))
max_iter = 500000
learning_rate = 0.1
weight = np.zeros(2)

for _ in range(max_iter):
    p = sig(np.dot(px_train, weight))
    weight = weight - (learning_rate / (n1 + n2)) * np.matmul(px_train.T, (p - y_train))

print(weight)
```

```python
x_test = pca.transform(x_test)
print(x_test)
x_test = np.dot(omega, x_test.T)
# px_test = [np.array([1]+list(i)) for i in x_test]
# px_test = np.array(px_test)
```

```python
x_t = [[1, i] for i in x_test]
x_t = np.array(x_t)
x_t
```

```python
y_pred = sig(np.dot(x_t, weight))
y_final = [1 if i > 0.5 else 0 for i in y_pred]
# y_final
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_final))
```

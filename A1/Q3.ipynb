```python
import csv

X = []
Y = []

with open('Real estate.csv', 'r') as file:
    reader = csv.reader(file)
    for row in reader:
        X.append([1.0]+[float(val) for val in row[1:-1]])
        Y.append([float(row[-1])])

print(X)
print(Y)
```

```python
x = np.array(X)
y = np.array(Y)
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
theta = np.linalg.inv(x_train.T.dot(x_train)).dot(x_train.T).dot(y_train)
```

```python
theta
```

```python
y_pred = x_test.dot(theta)
y_pred
```

```python
y_mean = y_pred.mean()
y_mean
```

```python
tss = np.sum((y_test-y_mean).T.dot(y_test-y_mean))
tss
```

```python
rss = np.sum((y_test-y_pred).T.dot(y_test-y_pred))
rss
```

```python
r2 = 1 - rss/tss
r2
```

```python
rmse = np.sqrt(np.sum(np.square(y_test-y_pred)/(len(y_test))))
rmse
```

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('glass.csv')
df
```

```python
mew = df.drop(columns=['Type']).mean()
mew = mew.to_numpy()
print(mew)
```

```python
cov_inv = np.linalg.inv(np.cov(df.drop(columns=['Type']).to_numpy().T))
cov_inv
```

```python
x = df.drop(columns=['Type']).to_numpy()
x
```

```python
print((x-mew).shape)
print((x-mew).T.shape)
print(cov_inv.shape)
```

```python
print((x[0]-mew).reshape(1,-1).shape)
print((x[0]-mew).reshape(-1,1).shape)
print(cov_inv.shape)
```

```python
distance = []
for i in x:
    x_ = (i-mew).reshape(9,1)
    d = np.sqrt(np.dot(np.dot(x_.T, cov_inv), x_))
    distance.append(d[0])
distance = np.array(distance)
distance
```

```python
import math
distance_thresholds = np.arange(min(distance)+0.1, max(distance)-0.1, 0.1)
# print(distance_thresholds)

min_var = 10
min_threshold = 0

for t in distance_thresholds:
    inliers = []
    outliers = []

    for i in range(len(distance)):
        if distance[i] < t:
            inliers.append(distance[i])
        else:
            outliers.append(distance[i])

    w0 = len(inliers)/len(distance)
    w1 = len(outliers)/len(distance)

    sig0_2 = np.var(inliers)
    sig1_2 = np.var(outliers)
    sig_2 = w0 * sig0_2 + w1 * sig1_2
    if sig_2 < min_var:
        min_var = sig_2
        min_threshold = t

print(min_threshold, min_var)
```

```python
count = 0
for i in distance:
    if i > min_threshold:
        count += 1
```

```python
# Implementing local outlier factor
k = 3
distance_matrix = np.zeros((214,214))

for i in range(214):
    for j in range(214):
        distance_matrix[i][j] = np.sqrt(np.sum((x[i] - x[j])**2))#np.linalg.norm(x[i]-x[j])
distance_matrix
```

```python
np.sort(distance_matrix[-1])
```

```python
k_distance = []
knn = []
for i in range(214):
    knn.append(np.argsort(distance_matrix[i])[1:k+1])
    # print(knn)
    k_distance.append(distance_matrix[i][knn[i][-1]])

k_distance
```

```python
knn = np.array(knn)
knn
```

```python
reach_dist = np.zeros((214,214))
for i in range(214):
    for j in range(214):
        reach_dist[i][j] = max(k_distance[i], distance_matrix[i][j])
reach_dist
```

```python
lrd = [0 for i in range(214)]

for i in range(214):
    score = 0
    for j in knn[i]:
        score += reach_dist[j][i]
    lrd[i] = len(knn[i])/score
lrd
```

```python
lof = [0 for i in range(214)]

for i in range(214):
    score = 0
    for j in knn[i]:
        score += lrd[j]/lrd[i]
    lof[i] = score/len(knn[i])
lof
```

```python
distance_thresholds = np.arange(min(lof) + 0.1, max(lof) - 0.1, 0.1)
# print(distance_thresholds)

min_var = 10000
min_threshold = 0

for t in distance_thresholds:
    inliers = []
    outliers = []

    for i in range(len(lof)):
        if lof[i] < t:
            inliers.append(lof[i])
        else:
            outliers.append(lof[i])

    w0 = len(inliers) / len(lof)
    w1 = len(outliers) / len(lof)

    sig0_2 = np.var(inliers)
    sig1_2 = np.var(outliers)
    sig_2 = w0 * sig0_2 + w1 * sig1_2
    if sig_2 < min_var:
        min_var = sig_2
        min_threshold = t

print(min_threshold, min_var)
count = 0
for i in lof:
    if i > min_threshold:
        count += 1
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('Iris_Q5.csv')
df['Species'] = df['Species'].map({'Iris-setosa':1, 'Iris-versicolor':2, 'Iris-virginica':3})
```

```python
c2 = df[df.Species == 2]
x2 = c2.drop(columns=['Id', 'Species']).to_numpy()
y2 = c2['Species'].to_numpy()
```

```python
x2_train, x2_test, y2_train, y2_test = train_test_split(x2, y2, test_size=0.2, random_state=42)
```

```python
c3 = df[df.Species == 3]
x3 = c3.drop(columns=['Id', 'Species']).to_numpy()
y3 = c3['Species'].to_numpy()
```

```python
x3
```

```python
y3
```

```python
x3_train, x3_test, y3_train, y3_test = train_test_split(x3, y3, test_size=0.2, random_state=42)
```

```python
def sig(x):
    return 1/(1+np.exp(-x))
```

```python
fx1 = []
fy1 = []
for i in x1_train:
    fx1.append(i)
    fy1.append(1)

for i in x2_train:
    fx1.append(i)
    fy1.append(0)

for i in x3_train:
    fx1.append(i)
    fy1.append(0)

fx1 = np.array(fx1)
fy1 = np.array(fy1)
```

```python
print(fx1)
```

```python
# print(fy1)
fy1 = fy1.reshape(-1,1)
print(fy1.shape)
```

```python
weight_1 = np.zeros(4)
weight_1 = weight_1.reshape(-1, 1)
max_iter = 500000
learning_rate = 0.1
n1 = len(x1_train)
n2 = len(x2_train) + len(x3_train)

for _ in range(max_iter):
    p = sig(np.dot(fx1, weight_1))
    p = p.reshape(-1,1)
    weight_1 = weight_1 - (learning_rate / (n1 + n2)) * np.matmul(fx1.T, (p - fy1))

print(weight_1)
```

```python
fx2 = []
fy2 = []
for i in x1_train:
    fx2.append(i)
    fy2.append(0)

for i in x2_train:
    fx2.append(i)
    fy2.append(1)

for i in x3_train:
    fx2.append(i)
    fy2.append(0)

fx2 = np.array(fx2)
fy2 = np.array(fy2)
```

```python
fy2 = fy2.reshape(-1,1)
print(fy2)
```

```python
weight_2 = np.zeros(4)
weight_2 = weight_2.reshape(-1, 1)
max_iter = 500000
learning_rate = 0.1
n1 = len(x2)
n2 = len(x1) + len(x3)

for _ in range(max_iter):
    p = sig(np.dot(fx2, weight_2))
    p = p.reshape(-1,1)
    weight_2 = weight_2 - (learning_rate / (n1 + n2)) * np.matmul(fx2.T, (p - fy2))

print(weight_2)
```

```python
fx3 = []
fy3 = []
for i in x1_train:
    fx3.append(i)
    fy3.append(0)

for i in x2_train:
    fx3.append(i)
    fy3.append(0)

for i in x3_train:
    fx3.append(i)
    fy3.append(1)

fx3 = np.array(fx3)
fy3 = np.array(fy3)

# x_train3, x_test3, y_train3, y_test3 = train_test_split(fx3, fy3, test_size=0.1, random_state=42)
```

```python
fy3 = fy3.reshape(-1,1)
print(fy3)
```

```python
weight_3 = np.zeros(4)
weight_3 = weight_3.reshape(-1, 1)
max_iter = 500000
learning_rate = 0.1
n1 = len(x3)
n2 = len(x1) + len(x1)

for _ in range(max_iter):
    p = sig(np.dot(fx3, weight_3))
    p = p.reshape(-1,1)
    weight_3 = weight_3 - (learning_rate / (n1 + n2)) * np.matmul(fx3.T, (p - fy3))

print(weight_3)
```

```python
x_test = np.concatenate((x1_test, x2_test, x3_test))
y_test = np.concatenate((y1_test, y2_test, y3_test))
```

```python
x_test
```

```python
print(y_test)
```

```python
weights = np.concatenate((weight_1, weight_2, weight_3), axis=1)
```

```python
weights
```

```python
y_pred = sig(np.matmul(x_test, weights))
y_pred
```

```python
y_final = []
for i in y_pred:
    # print(i)
    p = np.argsort(i)
    # print(p)
    if p[-1] == 0:
        y_final.append(1)
    elif p[-1] == 1:
        y_final.append(2)
    else:
        y_final.append(3)
y_final
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_final))
```

In my opinion this model i.e. Logistic Regression provides better results than the KNN. I have manually tested both the models and LR is giving a perfect 100% percent accuracy most of the time while in case of KNN and KNN+LDA accuracy depends upon how the data is split in train and test data. Concisely we can say that the average accuracy of KNN model is less than the average accuracy of the LR model.

I have used 'one vs all' method of logistic regression for this problem. In this method we select each class one by one and rest are taken together. Now the problem is reduced into simple n logistic regression problems. Now we find the predicted value wrt to n weights we have obtained and for the weight we get the maximum value of the probability we will assign label respective to that weight to the test.
